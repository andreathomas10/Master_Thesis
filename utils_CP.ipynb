```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import beta
```

```python
plt.rcParams['text.usetex'] = True
alpha = 0.1
n1 =  100
n2 = 1000
n3 = 10000
l1 = np.floor((n1+1)*alpha)
a1 = n1 + 1 - l1
l2 =  np.floor((n2+1)*alpha)
a2 = n2 + 1 -l2
l3 =  np.floor((n3+1)*alpha)
a3 = n3 + 1 -l3


x = np.arange (0.01, 1, 0.0001)
y1 = beta.pdf(x,a1,l1)
y2 = beta.pdf(x,a2,l2)
y3 = beta.pdf(x,a3,l3)
sns.set_style("whitegrid")
plt.figure(figsize = (16,9))
plt.plot(x,y1, label = "n = 100", linewidth=5)
plt.plot(x,y2, label = "n = 1000", linewidth=5)
plt.plot(x,y3, label = "n = 10000", linewidth=5)
plt.xticks(np.arange(0.0, 1, 0.02), fontsize = 20)
plt.yticks(fontsize = 20)
plt.tick_params(labelleft = False)
plt.xlim(0.8, 1)
plt.ylim(-5, 140)
plt.vlines(x = 0.9,ymin = 0, ymax = 200, linestyle = "dashed", color = "grey", label = r'\textbf{$1-\alpha$}')
plt.title("Distribution of coverage (infinite test set)", fontsize = 24)
plt.legend(fontsize = 20)
```

# conditional coverage

```python
fig = plt.figure(figsize = (16,9))
ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
ax.bar([0, 1], [0, 100], 0.25, color = "blue", label = "Conditional Coverage")
ax.bar([0, 1.25], [0, 70], 0.25, color = "green", label = "Marginal Coverage")
ax.bar([0,2], [0, 100], 0.25, color = "blue")
ax.bar([0,2.25], [0, 130], 0.25, color = "green")
ax.bar([0,0], [0, 100], 0.25, color = "blue")
ax.bar([0,0.25], [0, 100], 0.25, color = "green")
ax.spines[['top', 'right']].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.set_xticks([0.125, 1.125, 2.125])
ax.set_xticklabels(['Marginal', 'Group 1', "Group 2"], fontsize = 20)
ax.set_xlim([-0.5, 2.5])
ax.set_yticks([])
ax.set_ylim([0, 150])
ax.set_ylabel("Miscoverage", fontsize = 20)
ax.set_title("Conditional Vs Marginal Coverage", fontsize = 24)
ax.hlines(y = 100, xmin = -10, xmax = 3, color = "red")

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=5, fontsize = 20)

plt.show()
```

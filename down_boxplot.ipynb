```python
import pandas as pd
import numpy as np
import seaborn as sns
import json 
import matplotlib.pyplot as plt
```

```python
# Specify the path to your JSON file
json_file_path = "/Users/andrea/Desktop/Thesis/down_scale_dict/"
names = ["camx_mape.json","camx_mse.json","full_boost_ind_mape.json", "full_boost_ind_mse.json", "full_rf_glo_mape.json", "full_rf_glo_mse.json",
         "full_rf_ind_mape.json","full_rf_ind_mse.json","full_ridge_glo_mape.json","full_ridge_glo_mse.json",
         "full_ridge_ind_mape.json","full_ridge_ind_mse.json", "part_boost_glo_mape.json", "part_boost_glo_mse.json",
         "part_boost_ind_mape.json","part_boost_ind_mse.json","part_rf_glo_mape.json","part_rf_glo_mse.json",
         "part_rf_ind_mape.json","part_rf_ind_mse.json","part_ridge_glo_mape.json", "part_ridge_glo_mse.json",
         "part_ridge_ind_mape.json","part_ridge_ind_mse.json", "full_boost_glo_mape.json","full_boost_glo_mse.json",
         "full_gp_mse.json","part_gp_mse.json", "part_gp_mape.json"]
dict_of_dict = {}
# Read and parse the JSON file
for name in names:
    with open(json_file_path + name, 'r') as json_file:
        dict_of_dict[name] = json.load(json_file)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize = (16,9))
labels = ["CAMx","RF","RF*","Boost","Boost*","GP", "GP*", "Ridge","Ridge*"]
positions =  [0,3,5,8,10,13,15, 18,20]
plt.boxplot([list(dict_of_dict["camx_mse.json"].values()),
            list(dict_of_dict["full_rf_glo_mse.json"].values()),
            list(dict_of_dict["part_rf_glo_mse.json"].values()),
            list(dict_of_dict["full_boost_glo_mse.json"].values()),
            list(dict_of_dict["part_boost_glo_mse.json"].values()),
            list(dict_of_dict["full_gp_mse.json"].values()),
            list(dict_of_dict["part_gp_mse.json"].values()),
            list(dict_of_dict["full_ridge_glo_mse.json"].values()),
            list(dict_of_dict["part_ridge_glo_mse.json"].values())],
            patch_artist=True,
            labels = labels,
            positions = positions)
plt.yscale("log")
plt.ylabel("log(MSE)", fontsize = 20)
plt.tick_params(axis='x', labelsize=24) 
plt.tick_params(axis='y', labelsize=24) 
plt.title("MSE for OA with global models", fontsize = 30)
```
